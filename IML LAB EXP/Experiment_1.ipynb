```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
n = len(x)

# Least Squares
b1 = (n*np.sum(x*y) - np.sum(x)*np.sum(y)) / (n*np.sum(x**2) - (np.sum(x))**2)
b0 = (np.sum(y) - b1*np.sum(x)) / n

y_pred = b0 + b1 * x

# Metrics
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)
adj_r2 = 1 - (1 - r2)*(n - 1)/(n - 1 - 1)
see = np.sqrt(np.sum((y - y_pred)**2) / (n - 2))

# Print answers
print("Regression Equation: y =", round(b1,2), "x +", round(b0,2))
print("MAE:", round(mae,2))
print("MSE:", round(mse,3))
print("RMSE:", round(rmse,2))
print("R2:", round(r2,2))
print("Adjusted R2:", round(adj_r2,2))
print("SEE:", round(see,2))

# Graphs
plt.figure(figsize=(12,5))

# Left graph
plt.subplot(1,2,1)
plt.scatter(x, y, label='Actual Data')
plt.plot(x, y_pred, label='Regression Line')
plt.title("Linear Regression with Data Points")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()

# Right graph
plt.subplot(1,2,2)
metrics = ['MAE','MSE','RMSE','SEE']
values = [mae, mse, rmse, see]
plt.bar(metrics, values)
plt.title("Error Metrics Comparison")
plt.xlabel("Metrics")
plt.ylabel("Error Value")

plt.tight_layout()
plt.show()
```
