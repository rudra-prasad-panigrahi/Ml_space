```python
import numpy as np
from sklearn.svm import SVC


X = np.array([[1,1],
              [2,1],
              [2,3],
              [3,3]])

y = np.array([1,1,-1,-1])


model = SVC(kernel='linear')
model.fit(X, y)


w = model.coef_[0]
b = model.intercept_[0]

print("Weight vector (w):", w)
print("Bias (b):", b)


y_pred = model.predict(X)
print("Predicted labels:", y_pred)
```

```python
#2.
import numpy as np


alpha = np.array([1,1,1])
y = np.array([1,-1,-1])


X = np.array([[0,1,1],
              [0,2,-1],
              [-1,0,2]])


w = np.sum(alpha[:,None] * y[:,None] * X, axis=0)

print("Weight vector (w):", w)


def predict(x):
    return np.sign(np.dot(w, x))


x_test = np.array([1,1,1])
print("Prediction:", predict(x_test))
```

```python
import numpy as np


alpha = np.array([1,1,1])
y = np.array([1,-1,-1])


X = np.array([[1,0,-1],
              [-1,2,0],
              [0,-1,1]])


w = np.sum(alpha[:,None] * y[:,None] * X, axis=0)

print("Weight vector (w):", w)


x = np.array([0.5, 1, -0.5])


f = np.dot(w, x)

print("f(x):", f)


prediction = np.sign(f)
print("Predicted class:", prediction)
```
