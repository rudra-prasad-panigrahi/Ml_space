```python
import numpy as np
import matplotlib.pyplot as plt


x = np.array([1, 2, 3]).reshape(-1,1)
y1 = np.array([2, 4, 6])
y2 = np.array([3, 5, 7])

def simple_lr_with_errors(x, y):

    Xb = np.hstack([np.ones((len(x),1)), x])


    beta = np.linalg.inv(Xb.T @ Xb) @ Xb.T @ y


    y_pred = Xb @ beta

    errors = y - y_pred

    SSE = np.sum(errors**2)
    MSE = SSE / len(y)
    RMSE = np.sqrt(MSE)
    MAE = np.mean(np.abs(errors))

    SS_tot = np.sum((y - np.mean(y))**2)
    R2 = 1 - (SSE / SS_tot)

    return beta, y_pred, errors, MAE, MSE, RMSE, SSE, R2

b1, y1_pred, err1, mae1, mse1, rmse1, sse1, r2_1 = simple_lr_with_errors(x, y1)
b2, y2_pred, err2, mae2, mse2, rmse2, sse2, r2_2 = simple_lr_with_errors(x, y2)

print("Y1 Metrics:", mae1, mse1, rmse1, r2_1)
print("Y2 Metrics:", mae2, mse2, rmse2, r2_2)


plt.figure(figsize=(10,10))


plt.subplot(2,2,1)
plt.scatter(y2, y2_pred, label="Y2 actual prediction")
plt.plot(y2, y2, color='red', label="Perfect Fit")
plt.title("Y2 - Perfect Fit")
plt.xlabel("Actual Y2")
plt.ylabel("Predicted Y2")
plt.legend()
plt.grid()

plt.subplot(2,2,2)
plt.hist(err2, bins=3)
plt.title("Residuals Histogram - Y2")

plt.subplot(2,2,3)
plt.scatter(y1, y1_pred, label="Y1 actual prediction")
plt.plot(y1, y1, color='red', label="Perfect Fit")
plt.title("Y1 - Perfect Fit")
plt.xlabel("Actual Y1")
plt.ylabel("Predicted Y1")
plt.legend()
plt.grid()


plt.subplot(2,2,4)
plt.hist(err1, bins=3)
plt.title("Residuals Histogram - Y1")

plt.tight_layout()
plt.show()
```
