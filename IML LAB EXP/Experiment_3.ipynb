```python
#1.
import numpy as np


X = np.array([[1,2],
              [1,3],
              [1,4]])

y = np.array([1,2,3])


lam = 1


I = np.eye(X.shape[1])


beta = np.linalg.inv(X.T @ X + lam * I) @ X.T @ y


y_pred = X @ beta

print("Beta (coefficients):", beta)
print("Predicted values:", y_pred)

print(f"\nRegression Equation:")
print(f"y = {beta[0]:.4f} + {beta[1]:.4f}x")
```

```python
import numpy as np


X = np.array([[1,2,1],
              [1,2,3],
              [1,3,4]])

y = np.array([1,2,3])


lam = 1

I = np.eye(X.shape[1])


beta = np.linalg.inv(X.T @ X + lam * I) @ X.T @ y


y_pred = X @ beta

print("Beta (coefficients):", beta)
print("Predicted values:", y_pred)

print(f"\nRegression Equation:")
print(f"y = {beta[0]:.4f} + {beta[1]:.4f}x1 + {beta[2]:.4f}x2")
```
