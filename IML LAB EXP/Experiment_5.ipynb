```python
import pandas as pd


data = {
    'weather': ['sunny', 'sunny', 'rainy', 'rainy'],
    'playgame': ['yes', 'yes', 'no', 'no']
}


df = pd.DataFrame(data)

print("Dataset:\n", df)


p_yes = len(df[df['playgame'] == 'yes']) / len(df)
p_no  = len(df[df['playgame'] == 'no']) / len(df)


p_sunny_given_yes = len(df[(df['weather'] == 'sunny') & (df['playgame'] == 'yes')]) / len(df[df['playgame'] == 'yes'])


p_sunny_given_no = len(df[(df['weather'] == 'sunny') & (df['playgame'] == 'no')]) / len(df[df['playgame'] == 'no'])


posterior_yes = p_yes * p_sunny_given_yes
posterior_no  = p_no * p_sunny_given_no

print("\nP(Yes | Sunny):", posterior_yes)
print("P(No  | Sunny):", posterior_no)


if posterior_yes > posterior_no:
    print("\nPrediction: YES (Game will be played)")
else:
    print("\nPrediction: NO (Game will not be played)")
```
